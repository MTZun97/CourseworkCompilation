### Figure 1

Use `data-fig1.npy` for this figure.

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

a   = np.load( 'data-fig1.npy' )

plt.figure()
plt.plot(a, color = "k", marker = "o")
plt.ylabel("Value")
plt.xlabel("Time")
plt.show();
```

___

### Figure 2

Use `data-fig2.npy` for this figure.

```python
b = np.load("data-fig2.npy")

plt.figure()
plt.plot(b[:,0], color = "r")
plt.plot(b[:,1], color = "g")
plt.plot(b[:,2], color = "b")

plt.xlabel("X value")
plt.ylabel("Y value")
plt.legend(["Red","Green","Blue"])
plt.show();
```

___

### Figure 3

Use `data-fig3.npy` for this figure.

```python
c = np.load("data-fig3.npy")

plt.figure()
plt.scatter(c[:,0],c[:,1], color = "b", alpha = 0.3, s = 200)

plt.ylabel("Y value")
plt.xlabel("X value")
plt.show();
```

___

### Figure 4

Use `data-fig4.npy` for this figure.

```python
d = np.load("data-fig4.npy")
d_x = range(0,5)

plt.figure()
plt.bar(d_x, d, color = "darkgrey")

xlabels = ["A", "B", "C", "D", "E"]

plt.ylabel("Value")
plt.xlabel("Category")
plt.xticks(ticks = d_x, labels = xlabels)
plt.show();
```

___

### Figure 5

Use `data-fig5.npy` for this figure.

*Hint*: Use 2D visualization and the "RdBu" colormap.

```python
e = np.load("data-fig5.npy")

plt.figure()
plt.pcolor(e, cmap = "RdBu", vmin = -1, vmax =1 )
plt.colorbar()
plt.show();
```

Use `data-fig-bonus-X.npy`, `data-fig-bonus-Y.npy` and `data-fig-bonus-Z.npy` for this figure.

```python
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm

f_x = np.load("data-fig-bonus-X.npy")
f_y = np.load("data-fig-bonus-Y.npy")
f_z = np.load("data-fig-bonus-Z.npy")

fig = plt.figure()
ax = fig.gca(projection='3d')
cset = ax.contour(f_x, f_y, f_z, extend3d=True, cmap=cm.coolwarm)
ax.clabel(cset, fontsize=9, inline=1)

plt.show();
```
