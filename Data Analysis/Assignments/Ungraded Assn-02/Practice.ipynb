```python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
```

```python
# Define ranges of random variables
x0 = np.linspace(-1, 2, 1001)
x1 = np.linspace(-4, 4, 1001)
x2 = np.linspace(0, 20, 1001)

#Use analytical equations to calculate the probability densities of all x values
pdf0 = stats.uniform.pdf(x0)
pdf1 = stats.norm.pdf(x1)
pdf2 = stats.chi2.pdf(x2, 4)

#Plot the PDFs
fig, ax = plt.subplots(1, 3, figsize = (14, 3))

ax[0].plot(x0, pdf0)
ax[1].plot(x1, pdf1)
ax[2].plot(x2, pdf2)


ax[0].set_title("Uniform")
ax[1].set_title("Normal")
ax[2].set_title("$\chi^2$ ($v$ = 4)")

plt.setp(ax, xlabel = "x")
plt.setp(ax, ylabel = "f(x)")

plt.show()
```

```python
# Define ranges of thresholds
u0 = np.linspace(-1, 2, 1001)
u1 = np.linspace(-4, 4, 1001)
u2 = np.linspace(0, 20, 1001)

# Use analytical equations to calculate the probability densities at all x values
sf0 = stats.uniform.sf(u0)
sf1 = stats.norm.sf(u1)
sf2 = stats.chi2.sf(u2, 4)

# Plot the survival functions
fig, ax = plt.subplots(1, 3, figsize = (14, 3))

ax[0].plot(u0, sf0)
ax[1].plot(u1, sf1)
ax[2].plot(u2, sf2)

ax[0].set_title("Uniform")
ax[1].set_title("Normal")
ax[2].set_title("$\chi^2$ ($v$ = 4 )")

plt.setp(ax, xlabel = "x")
plt.setp(ax, ylabel = "Survival Function")

[axx.axhline(0, color = "k", linestyle = ":") for axx in ax]
[axx.axhline(0, color = "k", linestyle = ":") for axx in ax]

plt.show()
```

```python
np.random.seed(0)

N     = 5

# analytical PDF:
x     = np.linspace(0, 20, 1001)  # random variable range for the chi2 distribution
pdf   = stats.chi2.pdf(x, N)      # PDF for a chi2 random variable (with degrees of freedom = N)



niter = 10000
ss    = []
for i in range(niter):
    y = np.random.randn(N)
    print("y",y)
    print("$\y^2$:",y**2)
    print("sum:",(y**2).sum())
    ss.append( (y**2).sum() )
ss    = np.array(ss)



plt.figure()
ax = plt.axes()
ax.plot(x, pdf, label='Analytical')
ax.hist(ss, density=True, bins=21, label='Numerical')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('Probability density')
plt.show()
```
